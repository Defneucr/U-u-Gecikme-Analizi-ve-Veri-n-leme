```python
x = df.groupby(["Sezon"])["Hava Durumu"].transform(
        lambda x:x.mode()[0]
)
df["Hava Durumu"]=df["Hava Durumu"].fillna(x)
df["Hava Durumu"].isnull().sum()
```

```python
x = df.groupby(["Sezon"])["Bakım Durumu"].transform(
        lambda x:x.mode()[0]
)
df["Bakım Durumu"]=df["Bakım Durumu"].fillna(x)
df["Bakım Durumu"].isnull().sum()
```

```python
x = df.groupby(["Bakım Durumu"])["Planlanan Süre"].transform(
        lambda x:x.mode()[0]
)
df["Planlanan Süre"]=df["Planlanan Süre"].fillna(x)
df["Planlanan Süre"].isnull().sum()
```

```python
x = df.groupby(["Hava Durumu"])["Yolcu Sayısı"].transform(
        lambda x:x.mode()[0]
)
df["Yolcu Sayısı"]=df["Yolcu Sayısı"].fillna(x)
df["Yolcu Sayısı"].isnull().sum()
```

```python
x = df.groupby(["Sezon","Hava Durumu"])["Bagaj Ağırlığı"].transform("median")
df["Bagaj Ağırlığı"]=df["Bagaj Ağırlığı"].fillna(x)
df["Bagaj Ağırlığı"].isnull().sum()
```

```python
df = pd.get_dummies(df,columns=["Havayolu","Kalkış Havalimanı","Yolcu Yoğunluğu"],
                   drop_first=True,dtype=int)
```

```python
df
```

```python
(df["Uçuş Tipi"])=df["Uçuş Tipi"].map({
    'İç Hat':0,
    'Dış Hat':1    
})
```

```python
(df["Hava Durumu"])=df["Hava Durumu"].map({  
    
    'Yağmurlu':3, 
    'Açık':1, 
    'Bulutlu':2, 
    'Fırtınalı':4, 
    'Karlı' :5
})
```

```python
(df["Uçak Yaşı"])=df["Uçak Yaşı"].map({
    
    'Çok Eski (20+ yıl)':4, 
    'Orta (6-10 yıl)':2, 
    'Yeni (0-5 yıl)':1,
    'Eski (11-20 yıl)':3
      
})
```

```python
(df["Bakım Durumu"])=df["Bakım Durumu"].map({
    'Kritik':4, 
    'Normal':3, 
    'İyi':2, 
    'Mükemmel':1, 
    'Dikkat':5
                                                   
})
```

```python
(df["Sezon"])=df["Sezon"].map({
    'İlkbahar':2,
    'Yaz':1, 
    'Sonbahar':3, 
    'Kış':4
    
})
```

```python
df
```

```python
sutunlar = ["Planlanan Süre","Yolcu Sayısı","Bagaj Ağırlığı","Gecikme Süresi" ]

for sutun in sutunlar:
    Q1 = df[sutun].quantile(0.25)                      
    Q3 = df[sutun].quantile(0.75)  

    IQR = Q3 - Q1

    alt_sinir =Q1 - 1.5*IQR
    ust_sinir =Q3 + 1.5*IQR

    df[sutun] = df[sutun].clip(lower=alt_sinir,upper=ust_sinir)
```

```python
for sutun in sutunlar:
    print(f"{sutun}----{df[sutun].skew()}")
```
